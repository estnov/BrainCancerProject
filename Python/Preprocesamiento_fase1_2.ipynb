<h4>Lectura y transformacion de las imagenes en 1 solo dataset</h4>

```python
import numpy as np 
import pandas as pd 
from PIL import Image

import pandas as pd

import os
```

```python
df = pd.DataFrame()
count = 0
maxVal = 4000

for dirname, _, filenames in os.walk("D:/Datasets/BrainTumor/Fase_1/ImagenesProcesadas"):
    
    for filename in filenames:
        if(count<maxVal):
            list = dirname.split("\\")
            tipo = (list[1])
            
            image_path = os.path.join(dirname, filename)
            image_path=image_path.replace('\\','/')

            image = Image.open(image_path)

            # Se convierte a una matriz de pixeles
            pixel_matrix = image.load()

            width, height = image.size

            pixel_values = []

            pixel_matrix = np.array(image)

            pixel_values = pixel_matrix.flatten()

            dfPix = pd.DataFrame(pixel_values.reshape(1, -1), columns=[f'Pixel_{i+1}' for i in range(len(pixel_values))])


            try:
                dfPix['class'] = tipo
                df = pd.concat([df, dfPix], axis=0)
                print("Saved")
                print(f"Loaded: {count*100/maxVal} %")
                count+=1
            except:
                print("error!!")
                print(dirname)
                print(pixel_values.shape)
                print(tipo)
                print()

            
        else:
            break
```
