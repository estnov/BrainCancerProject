```python
import numpy as np 
import pandas as pd 
from PIL import Image

import pandas as pd

import os
```

```python
df = pd.DataFrame()
count = 0
maxVal = 3200

for dirname, _, filenames in os.walk("C:/Users/esteb/OneDrive - Universidad Politecnica Salesiana/Papers/Cancer cerebral/Dataset pequeño/archive"):
    
    for filename in filenames:
        if(count<maxVal):
            list = dirname.split("\\")
            tipo = (list[2])
            # Load the JPG image using Pillow
            image_path = os.path.join(dirname, filename)
            image_path=image_path.replace('\\','/')

            image = Image.open(image_path)

            # Convert the image to grayscale if needed
            image = image.convert('L')

            resized_image = image.resize((256, 256))

            # Get the pixel data as a matrix
            pixel_matrix = resized_image.load()

            # Get the dimensions of the image
            width, height = resized_image.size

            # Create a list to store the pixel values
            pixel_values = []

            # Get the pixel data as a matrix
            pixel_matrix = np.array(resized_image)

            # Reshape the pixel matrix into a 1D array
            pixel_values = pixel_matrix.flatten()

            # Create a DataFrame from the pixel values
            dfPix = pd.DataFrame(pixel_values.reshape(1, -1), columns=[f'Pixel_{i+1}' for i in range(len(pixel_values))])


            try:
                dfPix['class'] = tipo
                df = pd.concat([df, dfPix], axis=0)
                print("Saved")
                print(f"Loaded: {count*100/maxVal} %")
                count+=1
            except:
                print("error!!")
                print(dirname)
                print(pixel_values.shape)
                print(tipo)
                print()

            
        else:
            break
```

<h4>Red neuronal</h4>
<hr>

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau
from sklearn.model_selection import train_test_split
```

```python
x_train=X_train.values
x_test=X_test.values
print(X_train.shape)
print(x_test.shape)
```

```python
print(x_train.shape[0] * x_train.shape[1])
```

```python
x_train=np.array(x_train)
x_test=np.array(x_test)
x_train_reshaped = x_train.reshape(x_train.shape[0], 256, 256, 1)
x_test_reshaped = x_test.reshape(x_test.shape[0], 256, 256, 1)
```

```python
num_classes = 10
input_shape = (256, 256, 1)

#iniciamos el modelo
model = Sequential()
#añadimos una capa convolucional con 32 filtros de 3x3
model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape = input_shape))
#añadimos otra capa convolucional con 32 filtros de 3x3
model.add(Conv2D(32, (3, 3), activation='relu'))
#añadimos una capa de pooling de 3x3
model.add(MaxPooling2D(pool_size=(2, 2)))
#este dropout desactiva el 25% de las conexiones entre las neuronas, lo cual mejora los resultados
model.add(Dropout(0.25))

#repetimos todas las capas otra vez
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

#repetimos todas las capas otra vez
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(3, 3)))
model.add(Dropout(0.25))

#añadimos una capa para convertir las matrices en un vector
model.add(Flatten())
#añadimos una capa con 512 neuronas
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
#en la última capa tiene que hacer tantas neuronas como clases haya, en este caso 10
model.add(Dense(num_classes, activation='softmax'))

#compilamos el modelo
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
```

```python
batch_size = 180
epochs = 3
h = model.fit(x_train_reshaped,y_train, batch_size=batch_size,
                              epochs = epochs, validation_data = (x_test_reshaped,y_test))
```

```python
y_pred = model.predict(x_test_reshaped)
```

```python
classes_y_pred=np.argmax(y_pred,axis=1)
test_acc_cnn = model.evaluate(x_test_reshaped, y_test, verbose=0)
acc_cnn=str(round(test_acc_cnn[1], 4))
print('\nCNN Accuracy: ',acc_cnn)
```

```python
test_acc_cnn
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracyCNN = accuracy_score(y_test, classes_y_pred)
precisionCNN = precision_score(y_test, classes_y_pred, average='macro')
recallCNN = recall_score(y_test, classes_y_pred, average='macro')
f1CNN = f1_score(y_test, classes_y_pred, average='macro')

print("Accuracy:", accuracyCNN)
print("Precision:", precisionCNN)
print("Recall:", recallCNN)
print("F1-score:", f1CNN)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, classes_y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

<hr>

```python
def predict(image_path):
    image = Image.open(image_path)
    image = image.convert('L')
    resized_image = image.resize((256, 256))
    pixel_matrix = resized_image.load()
    width, height = resized_image.size
    
    pixel_values = []
    pixel_matrix = np.array(resized_image)
    pixel_values = pixel_matrix.flatten()

    dfPix = pd.DataFrame(pixel_values.reshape(1, -1), columns=[f'Pixel_{i+1}' for i in range(len(pixel_values))])
    
    y_result = classifier.predict(dfPix)
    
    if (y_result==0):
        print("Glioma")
    elif (y_result==1):
        print("Meningioma")
    elif (y_result==2):
        print("Sin tumor")
    elif (y_result==3):
        print("Tumor en la pituitaria")
    
```

<p>Los tumores cerebrales son complejos. Hay muchas anomalías en el tamaño y la ubicación de los tumores cerebrales. Esto hace que sea realmente difícil comprender completamente la naturaleza del tumor. Además, se requiere un neurocirujano profesional para el análisis de resonancia magnética. A menudo, en los países en desarrollo, la falta de médicos capacitados y la falta de conocimiento sobre los tumores hace que generar informes de resonancia magnética sea realmente desafiante y lento. Debido a ello nos hemos propuesto a diseñar diferentes metodos que pudiesen leer dichas imagenes de manera eficiente y rapida. Los resultados obtenidos muestran una clara superioridad de la máquina de soporte vectorial, la cual es atribuible a que la red neuronal convolucional falta por optimizar sus hiperparámetros, el modelo empleado es uno genérico. Otro resultado adicional es la asombrosa velocidad de ejecución de la SVM, con un tiempo de 2 minutos aproximadamente, en demérito de su contendiente que tuvo 15 minutos de entrenamiento.
