```python
def predict(df):
    filename = "svc.pickle"
    with open(filename, 'rb') as file:
        model = pickle.load(file)
    predictions = model.predict(df)
    
    salida = []
    for pred in predictions:
        if(pred==1):
            salida.append("Positivo a c치ncer cerebral")
        else:
            salida.append("El diagn칩stico preliminar es negativo")
    
    return salida
```
